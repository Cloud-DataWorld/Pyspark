```python
from pyspark.sql.types import *

help(StructType)`
```

```python
schema = StructType().add(field = 'Series_reference',data_type=StringType(),nullable=True) \
                    .add(field = 'Period ',data_type=DateType(),nullable=True) \
                    .add(field = 'Data_value',data_type=StringType(),nullable=True) \
                    .add(field = 'Suppressed',data_type=StringType(),nullable=True) \
                    .add(field = 'STATUS',data_type=StringType(),nullable=True) \
                    .add(field = 'UNITS',data_type=StringType(),nullable=True) \
                    .add(field = 'Magnitude',data_type=IntegerType(),nullable=True) \
                    .add(field = 'Subject',data_type=StringType(),nullable=True) \
                    .add(field = 'Group',data_type=StringType(),nullable=True) \
                    .add(field = 'Series_title_1',data_type=StringType(),nullable=True) \
                    .add(field = 'Series_title_2',data_type=StringType(),nullable=True) \
                    .add(field = 'Series_title_3',data_type=StringType(),nullable=True) \
                    .add(field = 'Series_title_4',data_type=StringType(),nullable=True) \
                    .add(field = 'Series_title_5',data_type=StringType(),nullable=True) 
```

```python
df4 = spark.read.csv('C:\\sampledata\\business.csv',schema=schema,header=True)

df4.printSchema()
```
