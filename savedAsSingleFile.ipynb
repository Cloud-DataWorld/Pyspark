```python
import findspark

findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder.master("local").config("spark.sql.warehouse.dir", "file:///C:").getOrCreate()
```

```python
data = [(1,'srikanth',2000),(2,'manvith',3000),(3,'Akshainine',4000),(4,'Sandhya',5000)]

schema = ['id','name','sal']

df = spark.createDataFrame(data,schema)

df.show()
```

```python
df.write.csv('file:///C:/sampledata/singlefile/out.csv',header = True)

```
